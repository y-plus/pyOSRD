```python
from pyosrd import OSRD
from pyosrd.schedules import schedule_from_osrd


sim = OSRD(simulation='station_capacity2', dir='station_capacity2')
```

```python
sim.draw_infra_points()
```

```python
sim.space_time_chart(0)
```

```python
schedule_from_osrd(sim).plot()
```

```python
sim.reset_delays()
sim.add_delay('train0',100, 280)
```

```python
delayed = sim.delayed()
```

```python
delayed.space_time_chart(0)
```

```python
schedule_from_osrd(delayed).plot()
```
