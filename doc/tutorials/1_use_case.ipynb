## Loading a predefined infra

```python
OSRD.infras
```

```python
sim = OSRD(infra='c1', dir='tmp')

sim.draw_infra_points()
```

## Loading a predefined simulation

```python
OSRD.simulations
```

```python
sim = OSRD(simulation='c1_2trains', dir='tmp')
```

```python
print(sim.num_stations)
print(sim.trains)
print(sim.num_trains)
```

with plotly

```python
sim.space_time_chart_plotly(1, points_to_show=['switch','station', 'signal'], eco_or_base='base')
```

with matplotlib

```python
sim.space_time_chart(0)
```

Extract informations about the trains trajectories

```python
sim.points_encountered_by_train(0)
```

```python
sim.points_on_track_sections()
```

```python
sim.train_track_sections(0)
```

```python
sim.results
```
