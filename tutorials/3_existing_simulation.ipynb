# Tutorial pyOSRD #3
Use an existing simulation

```python
from pyosrd import OSRD
```

1. Create the simulation by indicating the folder containing `infra.json` and `simulation.json`

```python
sim = OSRD(
    dir='existing_simulation'
)
```

2. Inspect it

```python
sim.draw_infra_points()
```

```python
sim.num_stations, sim.track_section_lengths
```

```python
sim.trains, sim.num_trains
```

```python
sim.train_lengths
```

3. Run the simulation

```python
sim.run()
```

4. Inspect the results

```python
sim.space_time_chart_plotly(train=0, points_to_show=['station'])
```

```python
sim.space_time_chart_plotly(
    train=1,
    points_to_show=['departure', 'arrival', 'signal']
)
```

```python
sim.train_departure(0), sim.train_arrival(0)
```

```python
sim.points_encountered_by_train(0)
```
